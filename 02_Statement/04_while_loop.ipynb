```python
x=1
while(x<11):
    print("x is currently: ",x)
    print("adding 1 to x")
    x+=1
else:
    print("all done")
```

1. break: breaks out of the current enclosing loop
2. continue: goes to the closent enclosing loop
3. pass: does nothing at all(placeholder)

```python
x=1
while(x<11):
    print("x is currently: ",x)
    print("adding 1 to x")
    x+=1
    if x==4:
        print("breaking because x is equal to 4")
        break
```

```python
i=1
while i<=10:
    if i%2 !=0:
        i+=1
        continue
    print(i)
    i+=1
```
