#### write a program that prints the integer from 1 to 100 but for multiples of 3 it prints iitr instead of the number and for multuples of 5 it prints machine learning and for multiples of both 3 and 5v it prints your name

* for loop and if-elif-else

```python
for num in range(1,101):
    if num%3==0 and num%5==0:
        print("Kavya")
    elif num%3==0:
        print("IITR")
    elif num%5==0:
        print("machine learning")
    else:
        print(num)
```

#### sum all the numbers of 1 to N excluding multiples of 3 or 7

```python
N= int(input("enter your number here: "))
total=0
for num in range(1,N+1):
    if num%3==0 or num%7==0:
        continue
    total=total+num
print(total)
```

#### check whether a number is palindrome

```python
num= int(input("enter a number here: "))
original= num

reverse=0

while num>0:
    reverse= reverse*10 + num%10
    num= num//10

print(f"the number {original} is a palidrome" if original==reverse else "not a palidrome")
```
