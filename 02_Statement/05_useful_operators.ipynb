#### enumerate

```python
idx=0
for c in 'xfdgh':
    print(f'at index {idx} the letter is {c}')
    idx+=1
```

```python
for idx, c in enumerate['xfdgh']:
    print(f"at index {idx} the letter is {c}")
```

```python
for idx, c in enumerate('xfdgh'):
    print(f"at index {idx} the letter is {c}")
```

```python
list(enumerate('cgfhbqxj'))
```
