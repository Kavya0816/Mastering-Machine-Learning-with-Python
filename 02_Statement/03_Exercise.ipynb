#### given marks of students, categories them into:
* 'A': 90-100
* 'B': 80-89
* 'C':70-79
* 'D':60-69
* 'F':<60

```python
scores=[36,97,85,74,76,92,86,63,59,72,97,67,84]
grade={'A':[],
        'B':[],
        'C':[],
        'D':[],
        'F':[]
       }

for sc in scores:
    if sc>=90:
        grade['A'].append(sc)
    elif sc>=80:
        grade['B'].append(sc)
    elif sc>=70:
        grade['C'].append(sc)
    elif sc>=60:
        grade['D'].append(sc)
    else:
        grade['F'].append(sc)
print(grade)
```

#### generate a list of numbers from 1 to N where:
* 'even prime' for 2
* 'even' for even numbers
* 'odd primes' for odd primes
* 'odd' for odd numbers

```python
N= int(input("enter your number here:"))

for num in range(1,N+1):
    if num==2:
        print(f'{num} is even prime')
    elif num%2==0:
        print(f'{num} is even')
    else:
        is_prime= True #assuming prime initially
        for i in range(2,num):
            if num%i==0: #if divisible, not prime
                is_prime= False
                break
        if is_prime== True and num!=1: #1 is not prime
            print(f'{num} is odd prime')
        else:
            print(f'{num} is odd')
```

#### for any string, create a dictionary with two keys: vowels and consonents

```python
st= input("enter your string here:")

vowels=[]
consonants=[]

for character in st.lower():
    if character.isalpha():
        if character in 'aeiou':
            vowels.append(character)
        else:
            consonants.append(character)
print(f"the given string has the following vowels:{vowels}\n and the following consonants:{consonants}")
```
