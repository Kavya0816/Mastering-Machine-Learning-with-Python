### A few dictionary methods

```python
d={ 'key1':1,
   'key2':2,
   'key3':3}
```

```python
# method to return a list of all keys
d.keys()
```

```python
type(d.keys())
```

```python
list(d.keys()) # cast dictionary keys to standard python list
```

```python
len(d.keys())
```

```python
# method to return tuples of all items
d.items()
```

```python
list(d.items())
```
