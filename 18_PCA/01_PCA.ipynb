## Manual Implemetation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("cancer_tumor_data_features.csv")
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

sc= StandardScaler()
```

```python
sc_X= sc.fit_transform(df)
sc_X
```

```python
sc_X.mean()
```

```python
covariance_matrix= np.cov(sc_X,rowvar=False)
covariance_matrix
```

```python
np.linalg.eig(covariance_matrix) #this is a tuple of eigen values and eigen vectors, so well separate those out
```

```python
eigenvalues,eigenvectors= np.linalg.eig(covariance_matrix)
```

```python
eigenvalues
```

```python
eigenvectors
```

```python
num_of_components=2
```

```python
np.argsort(eigenvalues)
```

```python
sorted_keys= np.argsort(eigenvalues)[::-1][:num_of_components]
sorted_keys
```

```python
eigenvalues[sorted_keys]
```

```python
eigenvectors[:,sorted_keys]
```

```python
eigen_values, eigen_vectors= eigenvalues[sorted_keys], eigenvectors[:,sorted_keys]
```

```python
principal_components= np.dot(sc_X, eigen_vectors)
principal_components
```

```python
plt.figure(figsize=(8,6))
plt.scatter(principal_components[:,0],principal_components[:,1]) # this will give us the transformed data set
plt.ylabel("Second Principal Component")
plt.xlabel("First Principal Component")
```

### PCA with Scikit Learn

```python
from sklearn.decomposition import PCA
```

```python
pca= PCA(n_components=2)
```

```python
pc_sklearn= pca.fit_transform(sc_X) # principal component sklearn
pc_sklearn
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pc_sklearn[:,0],pc_sklearn[:,1]) # this will give us the transformed data set
plt.ylabel("Second Principal Component")
plt.xlabel("First Principal Component")
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dictionary= load_breast_cancer()
```

```python
cancer_dictionary.keys()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pc_sklearn[:,0],pc_sklearn[:,1],c= cancer_dictionary['target']) # this will give us the transformed data set
plt.ylabel("Second Principal Component")
plt.xlabel("First Principal Component")
```

### Model Attributes

```python
pca.n_components
```

```python
pca.components_
```

```python
pca.explained_variance_ratio_
```

```python
np.sum(pca.explained_variance_ratio_) #this is saying that we are getting 63 percent of variance using just two features
```

```python
pca_30= PCA(n_components=30) # now we are taking 30 features
```

```python
pca_30.fit(sc_X)
```

```python
pca_30.explained_variance_ratio_
```

```python
np.sum(pca_30.explained_variance_ratio_)
```

```python
explained_variance= []

for n in range(1,30):
    pca= PCA(n_components=n)
    pca.fit(sc_X)

    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1,30),explained_variance)
plt.xlabel("No of components")
plt.ylabel("Variance Explained")
```
