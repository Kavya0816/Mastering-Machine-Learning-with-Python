### Universal Array Functions

https://numpy.org/doc/stable/reference/ufuncs.html

```python
a = np.array([[1, 0],
              [0, 1]])
b = np.array([[4, 1],
              [2, 2]])
np.matmul(a, b)
```

#### Axis Logic
* axis0: vertical axis (rows)
* axis1: horizontal axis (columns)

```python
ranarr_2d = np.random.randint (0, 50, (3, 4)) 
ranarr_2d
```

```python
ranarr_2d.sum(axis=0)
```

```python
ranarr_2d.sum(axis=1)
```

```python
ranarr_2d.mean(axis=0)
```

```python
ranarr_2d.mean(axis=1)
```

```python
ranarr_2d.std(axis=0)
```

```python
ranarr_2d.std(axis=1)
```
