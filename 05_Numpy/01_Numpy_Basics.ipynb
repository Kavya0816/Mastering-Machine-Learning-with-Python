* Vector strictly 1-D Array
* Matrices: 2-D (matrix can strill have only 1 row or 1 column)

##### Why Numpy and not just a List:
1. Memory Efficient
2. Easily Expandable to N-dimensions
3. Better computational speed
4. Broadcasting Operations and Functions
