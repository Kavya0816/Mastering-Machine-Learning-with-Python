### Data is not copied. 'Slice of array' is just a view of original array. This avoid memory problem.

```python
# resetting the array again

arr= np.arange(11,40,3)
arr
```

```python
arr_copy= arr.copy() #to get a copy, need to be explicit

arr_copy
```

```python
slice_of_array=arr_copy[:5]
slice_of_array
```

```python
slice_of_array[:4]=90
```

```python
slice_of_array
```

```python
arr_copy
```

```python
arr
```
