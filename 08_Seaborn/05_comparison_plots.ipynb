```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('StudentsPerformance.csv')
df.head()
```

### jointplot

```python
sns.jointplot(data=df,x='math score',y='writing score') #here default is scatter(kind=scatter)
```

```python
sns.jointplot(data=df,x='math score',y='writing score',kind='hex')
```

```python
sns.jointplot(data=df,x='math score',y='writing score',kind='kde')
```

### pairplot

```python
sns.pairplot(data=df)
```

```python
sns.pairplot(data=df,hue='gender')
```

```python
sns.pairplot(data=df,hue='gender',corner=True)
```

```python
sns.pairplot(data=df,hue='gender',corner=True,palette='Blues')
```

```python
sns.pairplot(data=df,hue='gender',corner=True,palette='Blues',diag_kind='hist')
```
