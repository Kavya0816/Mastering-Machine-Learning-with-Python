```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('dm_office_sales.csv')
df.head()
```

### Rugplot

```python
plt.figure(figsize=(5,3))
#sns.rugplot(x='salary',data=df)
sns.rugplot(x='salary',data=df,height=0.5)
```

### Histogram

```python
#sns.displot(data=df,x='salary')
#sns.displot(data=df,x='salary',bins=10)
sns.displot(data=df,x='salary',bins=10,kde=True)
plt.show()
```

```python
sns.histplot(data=df,x='salary',bins=10,kde=True)
plt.show()
```

```python
sns.histplot(data=df,x='salary',bins=10,kde=True,lw=10) #lw= line width
plt.show()
```

```python
sns.histplot(data=df,x='salary',bins=10,kde=True,lw=10, edgecolor='red')
plt.show()
```

```python
sns.kdeplot(data=df,x='salary')
```
