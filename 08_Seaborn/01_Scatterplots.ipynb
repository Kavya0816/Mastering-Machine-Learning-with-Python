```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
df.info()
```

### Scatterplot

```python
sns.scatterplot(x='salay',y='sales',data=df)
```

```python
sns.scatterplot(x='salary',y='sales',data=df)
```

```python
plt.figure(figsize=(9,3))
sns.scatterplot(x='salary',y='sales',data=df)
```

```python
plt.figure(figsize=(9,3))
sns.scatterplot(x='salary',y='sales',data=df,hue='division') #hue has categorical values
plt.savefig('example.png')
```

https://matplotlib.org/stable/users/explain/colors/colormaps.html

```python
plt.figure(figsize=(9,3))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience',palette='Greys') #hue has categorical values
```

```python
plt.figure(figsize=(9,4))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience',palette='Set1',size='training level',style='division')
plt.legend(loc=(1.1,0.1))
plt.show()
```
