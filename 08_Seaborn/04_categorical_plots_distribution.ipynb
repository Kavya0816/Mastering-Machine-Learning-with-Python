```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('StudentsPerformance.csv')
df.head()
```

### Boxplot

```python
plt.figure(figsize=(12,5))
#sns.boxplot(data=df,x='parental level of education',y='reading score')
sns.boxplot(data=df,x='parental level of education',y='reading score', hue='gender')
plt.legend(loc=(1.1,0.8))
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(data=df,y='parental level of education',x='reading score', hue='gender',orient='h')
plt.legend(loc=(1.1,0.8))
plt.show()
```

### Violinplot

```python
plt.figure(figsize=(12,5))
sns.violinplot(data=df,x='parental level of education',y='reading score')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.violinplot(data=df,x='parental level of education',y='reading score', hue='gender')
plt.legend(loc=(1.1,0.8))
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.violinplot(data=df,x='parental level of education',y='reading score', hue='gender',split=True) #since gender(hue) is a binary category hence we can use split for better results
plt.legend(loc=(1.1,0.8))
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.violinplot(data=df,x='parental level of education',y='reading score', hue='gender',split=True,inner='quart') # inner default is box
plt.legend(loc=(1.1,0.8))
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.violinplot(data=df,x='parental level of education',y='reading score', hue='gender',split=True,inner=None)
plt.legend(loc=(1.1,0.8))
plt.show()
```

```python
plt.figure(figsize=(5,12))
sns.violinplot(data=df,y='parental level of education',x='reading score', hue='gender',split=True,inner=None)
plt.legend(loc=(1.1,0.8))
plt.show()
```

### Swarmplots

```python
sns.swarmplot(data=df,x='math score')
```

```python
sns.swarmplot(data=df,x='math score',size=4)
```

```python
sns.swarmplot(data=df,x='math score',size=4,y='race/ethnicity')
```

```python
sns.swarmplot(data=df,x='math score',size=3,y='race/ethnicity',hue='gender', dodge=True)
```
