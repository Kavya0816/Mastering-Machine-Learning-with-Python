```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
blobs= pd.read_csv('cluster_blobs.csv')
blobs.head()
```

```python
sns.scatterplot(data= blobs,x='X1',y='X2')
```

```python
moons= pd.read_csv('cluster_moons.csv')
```

```python
moons.head()
```

```python
sns.scatterplot(data=moons,x='X1',y='X2')
```

```python
circles= pd.read_csv('cluster_circles.csv')
circles.head()
```

```python
sns.scatterplot(data=circles,x='X1',y='X2')
```

```python
def find_labels(model,data):
    labels= model.fit_predict(data)
    sns.scatterplot(data=data,x='X1',y='X2',hue=labels,palette='Set1')
```

```python
from sklearn.cluster import KMeans
```

```python
km_model= KMeans(n_clusters=2)
find_labels(model=km_model,data=moons)
```

```python
km_model= KMeans(n_clusters=3)
find_labels(model=km_model,data=blobs)
```

```python
km_model= KMeans(n_clusters=2)
find_labels(model=km_model,data=circles)
```

### DBSCAN Results

```python
from sklearn.cluster import DBSCAN
```

```python
db_model= DBSCAN(eps=0.6) #try to increase 0.6 to 0.7,0.8,0.9,0.99 and see that the no. of outliers is decreasing
find_labels(model=db_model,data=blobs)
```

```python
db_model= DBSCAN(eps=0.2) # put 0.0002 and see that it shows -1 in hue, this means all points are outliers now, and if we put 200 then it shows 0 as hue, this means everything belongs to one single cluster
find_labels(model=db_model,data=moons)
```

```python
db_model= DBSCAN(eps=0.2)
find_labels(model=db_model,data=circles)
```
