# Apply method

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('tips.csv')
df.head()
```

```python
'*'*12+str(3560325168603410)[-4:]
```

```python
def last_four_digits(num):
    return '*'*12+str(num)[-4:]
```

```python
last_four_digits(123456789123456789)
```

```python
df['CC Number']
```

```python
df['masked_CC']= df['CC Number'].apply(last_four_digits)
```

```python
df.head()
```

```python
df['CC Number'].apply(lambda num: '*'*12+str(num)[-4:])
```

```python
df['final_bill']= df['total_bill'].apply(lambda amt: np.round(1.05*amt,2))
```
