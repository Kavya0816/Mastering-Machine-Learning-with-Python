```python
X= dataset.iloc[:,:-1].values # all rows and all column except the last column
y= dataset.iloc[:,-1].values # all rows of last column
```

```python
y
```

### Splitting the data into Training set and Test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=1/3,random_state=25)
```

### Training the simple linear regression model on the training set

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor= LinearRegression()
regressor.fit(X_train,y_train)
```

### Predicting the test set results

```python
y_pred= regressor.predict(X_test)
y_pred
```

### Visualising the training set result

```python
plt.scatter(X_train,y_train,color='Red') #scatter plot showing actualy vales that is used to train the model
plt.plot(X_train,regressor.predict(X_train),color='green') #line showing the model
plt.title('Salary VS Experience(Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
