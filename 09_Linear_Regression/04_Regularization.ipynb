```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('Advertising.csv')
df.head()
```

```python
X= df.drop('sales',axis=1)
y=df['sales']
```

### Polynomial Conversion

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
p_converter= PolynomialFeatures(degree=3,include_bias=False)
```

```python
p_features= p_converter.fit_transform(X)
```

```python
p_features.shape
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(p_features, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler()
```

```python
sc.fit(X_train)
```

```python
sc_X_train= sc.transform(X_train) #scalled x train
sc_X_test= sc.transform(X_test)
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
r_model= Ridge(alpha=10)
```

```python
r_model.fit(sc_X_train,y_train)
```

```python
r_model.predict(sc_X_test)
```

```python
r_test_pred= r_model.predict(sc_X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE= mean_absolute_error(y_test, r_test_pred)
```

```python
MSE= mean_squared_error(y_test,r_test_pred)
```

```python
MAE
```

```python
MSE
```

```python
RMSE= np.sqrt(MSE)
```

```python
RMSE
```

```python
MAE_cv= mean_absolute_error(y_test, r_cv_test_pred)
```

```python
MSE_cv= mean_squared_error(y_test,r_cv_test_pred)
```

```python
RMSE_cv= np.sqrt(MSE_cv)
```

### Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import ElasticNet
```

```python
param_grid={
    'alpha': [0.1,1.5,10,50,100],
    'l1_ratio': [0.1,0.5,0.7,0.9,0.95,0.99,1]
}
```

```python
base_elastic_model= ElasticNet()
```

```python
grid_model= GridSearchCV(estimator= base_elastic_model,
             param_grid= param_grid,
             cv=5)
```

```python
grid_model.fit(sc_X_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.predict(sc_X_test)
```

```python
grid_pred= grid_model.predict(sc_X_test)
```

```python
np.sqrt(mean_squared_error(y_test,grid_pred))
```
