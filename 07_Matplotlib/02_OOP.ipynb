```python
a= np.linspace(0,10,11)
b=a**2
```

```python
x=np.arange(0,10)
y=2*x
```

#### Creating a figure

```python
type(fig)
```

```python
#create blank canvas
fig=plt.figure()

#add set of axes to figure
axes= fig.add_axes([0,0,1,1]) # left, bottom with height (0 to 1)

#plot on that set of axes
axes.plot(x,y)

axes.plot(a,b)

plt.show()
```

### adding another set of axes to the figure

```python
fig= plt.figure()

axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.2,0.5,0.5])
axes3=fig.add_axes([1.1,0.5,0.3,0.5])

axes1.plot(x,y)
axes1.plot(a,b)
axes2.plot(x,y)
axes2.plot(b,a)
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('Bigger fig')
axes2.set_title('Smaller fig')
axes2.set_xlim(20,80)

axes3.plot(x,y)

plt.show()
```
