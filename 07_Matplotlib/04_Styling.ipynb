```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,10)
y= 2*x
```

### Legends

```python
fig= plt.figure(figsize=(4,3))
ax= fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x squared',ls='-.')
ax.plot(x,x**3,label='x cube',marker='*',ms=10,mfc='red', mec='blue', lw = 3, color='#D27171' )

# plt.legend(loc=4) # loc 4 means lower right location
plt.legend(loc='center')
plt.show()
```
