### Plotting on axes objects

```python
fig, axes= plt.subplots(nrows=1,ncols=2)

axes[0].plot(a,b)

axes[1].plot(b,a)
```

```python
fig, axes= plt.subplots(nrows=2,ncols=2, figsize=(12,12))

axes[0,0].plot(a,b)
axes[0,0].set_xlabel('xlabel of a vs b')

axes[-1,-2].plot(x,y)
axes[-1,-2].set_title('Title of x vs y')

axes[1,1].plot(b,a)

fig.suptitle('Figure Level Title') # super title
fig.supxlabel('Super X Label')
fig.supylabel('Super Y Label')

fig.subplots_adjust(wspace=0.9, hspace=0.5)

plt.show()
```
