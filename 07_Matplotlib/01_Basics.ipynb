```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,10)
x
```

```python
y=2*x
y
```

#### Basic MatplotLib Commands

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
y=x**2
y
```

```python
plt.plot(x,y)
plt.xlabel('X axis title')
plt.ylabel('y axis title')
plt.title('Title of the plot below')
plt.ylim(20,65) #it fixes a limit for y axis data
plt.xlim(4,7)
plt.savefig('example.png')
plt.show()
```
