* filter: return an iterator yielding those items of iterable for which function is true

```python
# map
def sq_root(num):
    sq_rt= num**0.5
    return sq_rt
```

```python
mynums=[1,2,3,4,5,6,7]
```

```python
sq_root(mynums)
```

```python
output=[]

for n in mynums:
    output.append(sq_root(n))
```

```python
def operate_string(my_string):
    if len(my_string)%2==0:
        return f"the string '{my_string}' has even length"
    else:
        return f"the string '{my_string}' has {len(my_string)} characters"
```

```python
map(operate_string,['john','sarah','keli','mike'])
```

```python
list(map(operate_string,['john','sarah','keli','mike']))
```

#### write a function that takes a two word string and return True if both the words begin with same letter

```python
def string_checker(mystr):
    if len(mystr.split())!=2:
        return 'please enter two worded string only'
    else:
        splited_str= mystr.lower().split()
        return splited_str[0][0]== splited_str[1][0] #str[1][0], here 1 means 2nd word, and 0 means first letter of that word
```

```python
string_checker(mystr='hello world')
```

```python
string_checker(mystr='hello hworld')
```

```python
string_checker(mystr='hello world fgh cfgvhbj')
```

```python
string_checker(mystr='crazy Chimpanzi')
```

#### capitalize the 1st and 4th letter of the name

```python
def demo_func(name):
    if len(name)>3:
        return name[:3].capitalize() + name[3:].capitalize()
    else:
        return 'the name is too short'
```

#### given a list of integer, it returns True if 3 appears next to another 3
* check_33([1,2,3,3,4,5])--> True
* check_33([1,2,4,5])--> False
* check_33([1,2,3,4,5,3])--> False

```python
def check_33(numbers):
    for i in range(len(numbers)-1):
        if numbers[i]==3 and numbers[i+1]==3:
            return True
    return False
```

```python
check_33([1,2,3,3,4,5])
```

```python
check_33([1,2,4,5])
```

```python
check_33([1,2,3,4,5,3])
```

```python
# map, filter, lambda in def
def check_num_palindrome(ls):
    return list(filter(lambda n: str(n)==str(n)[::-1],ls))
```

```python
check_num_palindrome(ls=[123,232,234,565,707,56,789,9009,1234])
```
