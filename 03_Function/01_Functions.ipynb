```python
# check if any number in a list is even

def check_even_list(num_list):
    for num in num_list:
        if num%2==0:
            return True
```

```python
check_even_list(num_list=[11,22,33,44,55,66,77,88,99]) #this will give true if there is atleast one even number in the list
```

```python
def check_even_list(num_list):
    for num in num_list:
        if num%2==0:
            return True
        else:
            return False
```

```python
def check_even_list(num_list):
    for num in num_list:
        if num%2==0:
            return True
        else:
            pass
    return False
```

```python
check_even_list(num_list=[1,3,5])
```

```python
# return all the even numbers in a list
def check_even_list(num_list):
    even_numbers=[]
    for num in num_list:
        if num%2==0:
            even_numbers.append(num)
        else:
            pass
            
    return even_numbers
```

```python
check_even_list(num_list=[1,3,5])
```

```python
check_even_list(num_list=[1,2,3,4,5,6,7,8,9])
```

```python
# functions with tuple unpacking

def best_employee(wh_list):
    # setting some max to beat
    current_max=0
    # setting employee string
    employee_award_winner=''

    for e,h in wh_list:
        if h>current_max:
            current_max=h
            employee_award_winner=e
        else:
            pass
    return f"the employee of the year award goes to {employee_award_winner} who clocked {current_max} hours"
```

```python
weekly_work_hours=[('Akash', 70), ('Bob', 9), ('Chetan', 74), ('ta', 82), ('Ira',99)]
```

```python
best_employee(wh_list=weekly_work_hours)
```
