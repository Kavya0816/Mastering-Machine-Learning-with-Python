```python
class Circle:
    pi=22/7 # class level attribute

    def __init__(self,radius=1):
        self.radius=radius
        self.area= radius*radius*Circle.pi #or self.pi

    # Method for resetting the radius
    def setRadius(self,new_radius):
        self.radius= new_radius
        self.area= new_radius*new_radius*Circle.pi

    def getCircumference(self):
        return self.radius*self.pi*2

    def getArea(self):
        return self.radius*self.radius*self.pi
```

#### line class that accepts coordinates as a tuple and return slope and distance

```python
class Line:
    def __init__(self,coor1,coor2):
        self.coor1= coor1
        self.coor2= coor2

    def distance(self):
        x1,y1= self.coor1
        x2,y2= self.coor2
        return ((x2-x1)**2 + (y2-y1)**2)**0.5

    def slope(self):
        x1,y1=self.coor1
        x2,y2= self.coor2
        return (y2-y1)/(x2-x1)
```

```python
c1=(5,6)
c2=(13,17)

l1=Line(coor1=c1,coor2=c2)
```

```python
l1.coor1
```

```python
l1.coor2
```

```python
l1.distance()
```

```python
l1.slope()
```

### Special Methods

```python
class Book:
    def __init__(self,title,pages,author):
        print('a book is created')
        self.title=title
        self.author=author
        self.pages=pages

    def __str__(self):
        return f"the book '{self.title}' is written by '{self.author}' has '{self.pages}' pages"

    def __len__(self):
        return self.pages

    def __del__(self):
        print("your book is destroyed")
```

```python
b1=Book(title="MyLife", author="Jane", pages=500)
```

```python
print(b1)
```

```python
len(b1)
```

```python
del b1
```

```python
print(b1)
```

#### Create a bank account that has two attributes: owner, balance
* Two methods: deposit and withdraw
* Instantiate the bank account, make deposits and withdrawals

```python
class BankAccount:
    def __init__(self,owner,balance):
        self.owner= owner
        self.balance= balance

    def __str__(self):
        return f"account owner: {self.owner} \naccount balance: ${self.balance}"

    def deposit(self,dep_amt):
        self.balance=self.balance+dep_amt
        print(f"deposit successful, update balance is {self.balance}")

    def withdraw(self,wd_amt):
        if self.balance >= wd_amt:
            self.balance -= wd_amt
            print(f"Withdrawal Successful. Updated Balance is ${self.balance}")
        else:
            print(f"Funds Unavailable!! Balance is ${self.balance}. You are short of ${wd_amt-self.balance}")
```

```python
sbi_acccount= BankAccount(owner="shreya", balance=1000)
```

```python
print(sbi_acccount)
```

```python
sbi_acccount.deposit(dep_amt=500)
```

```python
print(sbi_acccount)
```

```python
sbi_acccount.withdraw(wd_amt=1200)
```

```python
sbi_acccount.withdraw(wd_amt=1200)
```
